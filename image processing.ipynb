```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from PIL import ImageFilter
from PIL import ImageEnhance  
import cv2
```

```python
PIL.__version__
```

```python
image = Image.open('comic-fish-md.png')
```

```python
image.size
```

```python
size = (128,128)
```

```python
#Resize Image
```

```python
img = image.resize(size)
```

```python
img.size
```

```python
plt.imshow(img)
```

```python
#Rotate image
plt.imshow(img.rotate(90))
```

```python
img = img.rotate(90)
```

```python
arr = img.split()
```

```python
plt.imshow(arr[2])
```

```python
#Convert Grayscale
plt.imshow(img.convert('LA'))
```

```python
#crop image
width, height = img.size   # Get dimensions
left = width/4
top = height/4
right = 3 * width/4
bottom = 3 * height/4
plt.imshow(img.crop((left, top, right, bottom))) 
```

```python
#sharpen image
#https://pillow.readthedocs.io/en/stable/reference/ImageFilter.html
plt.imshow(img.filter(ImageFilter.SHARPEN))
```

```python
#blur image
#https://pillow.readthedocs.io/en/stable/reference/ImageFilter.html
plt.imshow(img.filter(ImageFilter.BLUR))
```

```python
#smooth image
#https://pillow.readthedocs.io/en/stable/reference/ImageFilter.html
plt.imshow(img.filter(ImageFilter.SMOOTH))
```

```python
#edge enhance image
#https://pillow.readthedocs.io/en/stable/reference/ImageFilter.html
plt.imshow(img.filter(ImageFilter.EDGE_ENHANCE))
```

```python
#detail image
#https://pillow.readthedocs.io/en/stable/reference/ImageFilter.html
plt.imshow(img.filter(ImageFilter.DETAIL))
```

```python
#emboss image
#https://pillow.readthedocs.io/en/stable/reference/ImageFilter.html
plt.imshow(img.filter(ImageFilter.EMBOSS))
```

```python
#detail image
plt.imshow(img.filter(ImageFilter.DETAIL))
```

```python
#find edges image
#https://pillow.readthedocs.io/en/stable/reference/ImageFilter.html
plt.imshow(img.filter(ImageFilter.FIND_EDGES))
```

```python
#contrast image
enh = ImageEnhance.Contrast(img)  
plt.imshow(enh.enhance(1.8))
```

```python
#contrast image
#https://pillow.readthedocs.io/en/3.0.x/reference/ImageEnhance.html
enh = ImageEnhance.Contrast(img)  
plt.imshow(enh.enhance(1.8))
```

```python
#Brightness 
#https://pillow.readthedocs.io/en/3.0.x/reference/ImageEnhance.html
enh = ImageEnhance.Brightness(img)  
plt.imshow(enh.enhance(1.8))
```
